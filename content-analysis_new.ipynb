```python

import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_termes = pd.read_csv('output/df_termes.csv', sep='|')
df_views = pd.read_csv('output/df_views.csv', sep='|')
selected_websites = [
    "https://www.droit-compta-gestion.fr",
    "https://www.entreprendre-maintenant.fr"
]
output_dir = 'output'
```

```python
df_termes.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
df_termes
```

```python
my_topics = set(df_termes[df_termes['website'].isin(selected_websites)]['terme'])
```

```python
site_topics = df_termes[~df_termes['terme'].isin(my_topics)]
```

```python
site_topics = df_termes[~df_termes['website'].isin(['https://www.magazine-paris-berlin.com'])]
```

```python
site_topics
```

```python
site_topics_views = site_topics.merge(df_views, on='article_url', how='inner')
```

```python
site_topics_views = site_topics_views[site_topics_views['frequence'] > 5]
```

```python
site_topics_views.groupby('terme').mean('article_views').sort_values(by='article_views', ascending=False)
```

```python
%matplotlib inline
```

```python
text_data = ' '.join(
    site_topics_views['terme']
    .fillna('')        # remplace les NaN par vide
    .astype(str)       # convertit tout en string
)
wc = WordCloud(width=800, height=400, background_color='white').generate(text_data)
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
